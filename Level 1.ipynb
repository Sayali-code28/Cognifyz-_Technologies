```python
# Handle missing values based on column 
```

```python
# Missing numerical values with the mean
```

```python
num_columns=dataset.select_dtypes(include=['float64','int64']).columns
```

```python
dataset[num_columns] = dataset[num_columns].fillna(dataset[num_columns].mean())
```

```python
# Missing categorical values with the mode
```

```python
cat_columns= dataset.select_dtypes(include=['object']).columns
```

```python
dataset[cat_columns] = dataset[cat_columns].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
print("\nMissing Values handled successfully.")
```

```python
print(dataset.isnull().sum())
```

```python
# Perform data type conversion if necessary.
```

```python
if dataset['Aggregate rating'].dtype !='float64': dataset['Aggregate rating']= pd.to_numeric(dataset['Aggregate rating'],errors='coerce')
print(dataset.dtypes)
```

# Task 3 -Geospotial Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("Dataset .csv")
dataset.head(5)
```

```python
location_dataset = dataset[['Restaurant ID', 'Latitude', 'Longitude']].dropna()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x=location_dataset['Longitude'], 
    y=location_dataset['Latitude'], 
    hue=location_dataset['Restaurant ID'], 
    palette='tab10', 
    legend=False
)
plt.title('Restaurant Locations Based on Latitude and Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```
